```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("air_bnb_data.csv")
```

```python
df
```

```python
#Check the column names in the Dataset
df.columns
```

```python
#Check for missing values
print(df.isnull().sum())
```

```python
#Handle Missing Values
#Convert 'last review' to datetime and handle errors
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')

#Fill missing values
df.fillna({'reviews per month': 0, 'last review': df['last review'].min()}, inplace = True)

#Drop records with missing 'name' or 'host name'
df.dropna(subset=['NAME', 'host name'], inplace = True)
```

```python
print(df.isnull().sum())
```

```python
df['price'] = df['price'].str.replace('[\\$,]', '', regex=True).astype(float)
df['service fee'] = df['service fee'].str.replace('[\\$,]', '', regex=True).astype(float)

#Verify the changes
print(df[['price', 'service fee']].dtypes)
print(df[['price', 'service fee']].head())
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.info())
```

```python
df = df.drop(columns=["license", "house_rules"], errors='ignore') 
df
```

```python
df.describe()
```

```python
#visualization
#To plot the distribution of listing prices. 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.histplot(df['price'], bins=50, kde=True, color='blue') 
plt.title('Distribution of Listing Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
This histogram shows a fairly even distribution of listing prices across different price 
ranges indicating no particular concentration of listings in any specific price range. 
    The KDE line helps visualize this even spread more clearly, confirming that the dataset
conatins listings with a wide variety of prices.
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='room type', data=df, color='hotpink')
plt.title('Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

```python
The count plot shows a clear distribution of the different room types available 
in the Airbnb dataset. The majority of listings are for 'Entire home/apt' and 'Private room'
with 'Shared room' and 'Hotel room' being much less common. This insight can be useful for 
understanding the availability and popularity of different types of accomodations on Airbnb.
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(y = 'neighbourhood group', data=df, color="green", order=df['neighbourhood group'])
plt.title('Number of Listings by Neighbourhood Group')
plt.xlabel('Count')
plt.ylabel('Neighbourhood Group')
plt.show()
```

```python
The count plot shows a clear distribution of the number of listings across different 
neighbourhood groups. Manhattan and Brroklyn dominate the listings , suggesting they are prime
locations for Airbnb. Queens, Bronx, and Staten Island have fewer listings, indicating 
less availability or popularity.
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='room type', y = 'price', hue ='room type', data=df, palette='Set1')
plt.title('Price vs. Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price ($)')
plt.show()
```

```python
Price vs. Room Type
The box plot provides a detailed view of how prices vary across different room types in the
Airbnb dataset. It shows that while 'Shared room' tends to have lower prices, 'Private room',
'Entire home/apt' and 'Hotel room' have higher and more varied price ranges.
This visulaization helps in understanding the pricing dynamics for 
different types of accomodations on Airbnb.
```

```python
#Reviews Over Time
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12, 6))
reviews_over_time.plot(kind='line', color='red')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
This line plot provides a clear visualization of the number of reviews over time. 
    It helps identify trends and patterns in review activity, such as periods of high or low 
activity. This information can be useful for understanding the dynamics of user enagagement 
and the popularity of Airbnb listings over time. The significant spikes and drops
on reviews might be worth further investigation to understand the underlying causes, such as 
changes in Airbnb policies, market conditions, or external events.
```
